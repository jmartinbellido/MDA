---

### Utilizando Python base
---
La sintaxis b√°sica en Python para seleccionar elementos dentro de una estructura de datos (cualquier) es la siguiente:

```
data_structure[x]
```
* Utilizamos siempre brackets [ ] para seleccionar data almacenada
* Esto aplica tanto a *lists, dictionaries, tupples, sets y DataFrames*
* El tratamiento dentro de los brackets depende del tipo de estructura

> ‚ö†Ô∏è Revisar *notebook 1: Python Base*, en caso de dudas


#### Seleccionar columnas

Para seleccionar una columna, incluimos un objeto de tipo texto dentro del par√°metro de selecci√≥n,
```
df["column_name"]
```

En caso de querer seleccionar m√°s de una columna, incorporamos una list como par√°metro,

```
df[["column_1","column_2"]]
```





````
# Invocamos una columna espec√≠fica
df_cars["brand"]
````

````
# Invocamos tres columnas
df_cars[["brand","price","mileage"]]
````

De forma alternativa, cuando queremos seleccionar una √∫nica columna, podemos optar por la siguiente sintaxis simplificada,

```
object.column_name
```
> ‚úÖ Pandas interpreta los nombres de columnas como atributos del objeto



````
df_cars.brand
````

#### Seleccionar filas
Para seleccionar filas utilizando Python base, podemos incorporar como par√°metro un vector (*list*) con objetos *booleanos* (uno par cada fila en el DataFrame, en orden). Las filas que corresponsan a valores *True* quedar√°n seleccionadas.

```
df[[True,False,False...]]
```
> ‚úÖ En caso de detectar valores booleanos dentro del par√°metro de selecci√≥n, Python interpreta que estamos seleccionando filas (no columnas)

> ‚ö†Ô∏è El vector booleano deber√° tener tantos elementos, como filas existan en el DataFrame

Por el momento, esta receta no tiene mucho valor pr√°ctico, ya que no solemos filtrar filas creando manualmente vectores booleanos. No obstante, esto es fundamental para cuando aprendamos a filtrar filas seg√∫n condiciones.

````
# Creamos un list con valores booleanos (10 valores, tantos como filas en DataFrame)
# Utilizamos el list dentro del par√°metro de selecci√≥n para filtrar filas
my_boolean_list = [True,False,False,False,False,True,True,True,True,True]
df_cars[my_boolean_list]
# Como podemos comprobar, conservamos √∫nicamente las filas primera y √∫ltimas cinco, que coinciden con los valores True en el vector
````

### Filtrar filas, seg√∫n condici√≥n l√≥gica
---

Podemos utilizar operadores para crear vectores booleanos de forma din√°mica, luego utilizar estos para filtrar filas en un DataFrame.

```
object[object[condition]]
```



````
# Importamos df
df_jamesbond = pd.read_csv("https://data-wizards.s3.amazonaws.com/datasets/jamesbond.csv")
````

````
# A continuaci√≥n, seleccionaremos elementos del dataframe a partir de condiciones l√≥gicas
cond = df_jamesbond["Budget"]>100 # creamos un objeto con una prueba l√≥gica, esto nos devolver√° un vector booleano
cond
````

````
# Utilizaremos el objeto que almacena el vector booleano como par√°metro en el df
df_jamesbond[cond] # filtramos el df
````

````
# creamos condiciones y las almacenamos en objetos
cond_1 = df_jamesbond["Budget"]>100
cond_2 = df_jamesbond["Actor"]=='Daniel Craig'
cond_3 = df_jamesbond["Year"]<1965

df_jamesbond[(cond_1 & cond_2) | cond_3] # seleccionamos filas que cumplen con m√∫ltiples condiciones
````

````
# combinamos m√©todos para seleccinar filas y columnas
df_jamesbond[(cond_1 & cond_2) | cond_3][['Year','Director','Actor']]
````

````
# Tambi√©n podemos negar condiciones
cond_1 = df_jamesbond["Actor"]=='Daniel Craig'
df_jamesbond[-cond_1] # negamos nuestro vector booleano, convirtiendo los True en False y viceversa
````

#### Funciones √∫tiles para filtrar


````
# podemos utilizar el m√©todo .isin() para filtrar por m√∫ltiples valores en una variable categ√≥rica
cond = df_jamesbond['Director'].isin(['Martin Campbell','Terence Young']) # filtramos por m√°s de un valor para una variable categ√≥rica
df_jamesbond[cond]
````

````
# creamos condiciones para reconocer patrones espec√≠ficos de texto
cond_1 = df_jamesbond['Film'].str.contains('love', regex=False, case=False)
cond_2 = df_jamesbond['Film'].str.contains('die', regex=False, case=False)

# creamos una condici√≥n, reconociendo un patr√≥n en un campo de tipo texto
df_jamesbond[cond_1 | cond_2]
````

### Anidar operaciones
---
Python permite anidar operaciones con m√©todos de forma nativa. Es decir, podemos realizar m√∫ltiples operaciones sin necesidad de definir y sobrescribir variables.

Ejemplo:

```
df.iloc[x,y].head().sort_values(x)
```
* En este ejemplo, estar√≠amos realizando tres operaciones anidadas


````
# A modo de ejemplo, realizamos a continuaci√≥n cuatro operaciones anidadas
pd.read_csv("https://data-wizards.s3.amazonaws.com/datasets/jamesbond.csv").iloc[0:5,:].head().sort_values("Film")
````

#### Python method chaining
Al anidar m√∫ltiples operaciones, con frecuencia podemos encontrarnos realizando l√≠neas de c√≥digo demasiado extensas. Ante estas situaciones, podemos optar por separar nuestras operaciones en nuevas l√≠neas de c√≥digo utilizando una sintaxis especial llamada *method chaining*.

```
df.iloc[x,y]\
  .head()\
  .sort_values(x)
```
* utilizamos una barra invertida (\) para separar operaciones
* la barra invertida se coloca tras terminar una operaci√≥n
* tras la barra invertida, damos enter para cambiar de l√≠nea de c√≥digo; no debemos dejar ning√∫n caracter adicional

> ‚ö†Ô∏è *Method chaining no admite ning√∫n caracter tras la barra invertida, tampoco comentarios*

````
# Repetimos las cuatro operaciones anteriores, esta vez utilizando el method chaining
# Este m√©todo adquiere mayor utilidad cuanto m√°s operaciones realicemos
pd.read_csv("https://data-wizards.s3.amazonaws.com/datasets/jamesbond.csv")\
  .iloc[0:5,:]\
  .head()\
  .sort_values("Film")
````

##### Importar dataframe. Extraer top 10 pel√≠culas de Sci-Fi con mayor IMDB score, seleccionar √∫nicamente campos *title_year*, *director_name* y *imdb_score*.

> Dataset https://data-wizards.s3.amazonaws.com/datasets/movies.csv




````
df_movies = pd.read_csv('https://data-wizards.s3.amazonaws.com/datasets/movies.csv',index_col='movie_title')
df_movies.head()
````

````
# creamos una condici√≥n, reconociendo un patr√≥n en un campo de tipo texto
cond_1 = df_movies['genres'].str.contains('Sci-Fi', regex=False, case=False)

# filtramos por filas que cumplen la condici√±√≥n, ordenamos, seleccionamos primeras 10 entradas y seleccionamos determinadas columnas
df_movies[cond_1].sort_values('imdb_score',ascending=False).iloc[:10,][['title_year','director_name','imdb_score']]
````
